```python
#Criando gráfico básico linear (necessário instalar as libs matplotlib & numpy)

import matplotlib.pyplot as plt
import numpy as np
x = [0,1,2,3,4,5]
y = [0,1,2,3,4,5]
plt.plot(x, y, label='linear')
```

```python
#Gráfico com exemplos de gráficos exponenciais

import matplotlib.pyplot as plt
import numpy as np
# Gere 10 numeros entre 0-10
x = np.linspace(0, 10, 10)
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadrático')
plt.plot(x, x**3, label='cúbico')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title("Gráfico simples")
plt.legend()
plt.show()
```

```python
#Fazendo a média dos 3 gráficos anteriores
import matplotlib.pyplot as plt
import numpy as np
# Gere 10 numeros entre 0-10
x = np.linspace(0, 10, 10)
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadrático')
plt.plot(x, x**3, label='cúbico')

# Calculando a média aritmética dos 3 gráficos
media = (x + x**2 + x**3)/3

# Plotando o Gráfico da Média
plt.plot(x, media, label='média', linestyle = 'dotted')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title("Gráfico simples")
plt.legend()
plt.show()
```
